1. What is the difference between a function and a method in Python?
Function

A function is a block of reusable code that performs a specific task.

Defined using the def keyword.

Does not require an object to call.

Called directly by name.

Example:

def greet():
    print("Hello!")

Method

A method is a function that is defined inside a class.

A method always belongs to an object or class.

Must be called using an object.

Uses self to access object data.
| Function              | Method                 |
| --------------------- | ---------------------- |
| Defined independently | Defined inside a class |
| Called directly       | Called using an object |
| Does not use `self`   | Uses `self`            |
| Not tied to objects   | Always tied to objects |


```python
def greet():
    print("Hello!")

class Person:
    def greet(self):
        print("Hello from method!")

greet()

p = Person()
p.greet()
```

6. Explain the concept of generators in Python and how they are defined

Ans:- What Are Generators?

Generators in Python are special functions that produce values one at a time instead of returning all values at once.
They are memory-efficient and used for handling large data or streams of data.

Generators use the yield keyword instead of return.

Key Features of Generators:-

: Memory Efficient
They generate values only when needed (lazy evaluation).

: Return One Value at a Time
Unlike normal functions, generators pause after yielding a value.

: Use yield Instead of return
yield saves the function state and resumes on the next call.

: Automatically Create an Iterator
Generator functions return generator objects which behave like iterators.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

# Create generator object
gen = count_up_to(5)

# Use generator
for num in gen:
    print(num)
```

8. What is a lambda function in Python and when is it typically used?

Ans:- A lambda function in Python is a small, anonymous (nameless) function.
It is defined using the keyword lambda instead of def.

A lambda function can have any number of arguments but only one expression
Lambda functions are used when:

A short, simple function is needed for a short time.

You don’t want to formally define a function using def.

They are commonly used with:

map()

filter()

sorted()


```python
# 1. Basic lambda function
square = lambda x: x * x
print("Square of 5:", square(5))

# 2. Lambda with map() - doubling numbers
numbers = [1, 2, 3]
doubled = list(map(lambda x: x * 2, numbers))
print("Doubled Numbers:", doubled)

# 3. Lambda with filter() - even numbers
numbers2 = [1, 2, 3, 4, 5, 6]
even = list(filter(lambda x: x % 2 == 0, numbers2))
print("Even Numbers:", even)

# 4. Lambda with sorted() - sort by length
names = ["raj", "yuvraj", "aman"]
sorted_names = sorted(names, key=lambda x: len(x))
print("Sorted Names by Length:", sorted_names)
```

9. Explain the purpose and usage of the `map()` function in Python.

Ans:- The map() function in Python is used to apply a function to every item in an iterable (like a list, tuple, etc.) and returns a map object (an iterator) with the results.

It is useful when you want to perform the same operation on all elements of a collection without writing a loop.


```python
# 1. Using map() with a regular function
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
squared_numbers = map(square, numbers)
print("Squared Numbers (regular function):", list(squared_numbers))

# 2. Using map() with a lambda function
numbers = [1, 2, 3, 4]
squared_numbers_lambda = map(lambda x: x**2, numbers)
print("Squared Numbers (lambda):", list(squared_numbers_lambda))

# 3. Using map() with multiple iterables
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
sum_result = map(lambda x, y: x + y, numbers1, numbers2)
print("Sum of two lists:", list(sum_result))
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
| **Function** | **Purpose**                                                                  | **Example**                                            |
| ------------ | ---------------------------------------------------------------------------- | ------------------------------------------------------ |
| `map()`      | Applies a function to **all items** in an iterable                           | `squared = map(lambda x: x**2, numbers)`               |
| `filter()`   | Returns items that **satisfy a condition**                                   | `even_numbers = filter(lambda x: x % 2 == 0, numbers)` |
| `reduce()`   | Reduces the iterable to a **single value** by applying a function repeatedly | `sum_numbers = reduce(lambda x, y: x + y, numbers)`    |






```python
from functools import reduce

# --- map() ---
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))   # squares each number
print("map() - Squares:", squared)

# --- filter() ---
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))  # only even numbers
print("filter() - Even numbers:", even_numbers)

# --- reduce() ---
sum_numbers = reduce(lambda x, y: x + y, numbers)  # sum all numbers
print("reduce() - Sum:", sum_numbers)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Print all prime numbers from 1 to 200
prime_numbers = []
for n in range(1, 200 + 1):
    if is_prime(n):
        prime_numbers.append(n)

print(prime_numbers)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_value = self.a + self.b
            self.a = self.b
            self.b = next_value
            self.count += 1
            return next_value

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

# Example usage:
for value in power_of_two(10):
    print(value)
```

7. Implement a generator function that reads a file line by line and yields each line as a string

```python
def read_file_lines(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()   # removes extra spaces/newline


# Example usage:
for line in read_file_lines("sample.txt"):
    print(line)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda ch: ch not in vowels, text))

# Example
input_string = "Hello Python Programming"
result = remove_vowels(input_string)
print("String without vowels:", result)
```
